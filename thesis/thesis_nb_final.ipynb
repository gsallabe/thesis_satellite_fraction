## 3) Addressing Spectroscopic Incompleteness
This is where I am going to do a separate satellite fraction analysis wherein I model the spectroscopic incompleteness in the model data. I am going to model this by using a Gaussian to model the "kicks" based on the scatter (I believe from [Tanaka et al. (2018)](http://arxiv.org/abs/1704.05988)) 
$$ \sigma = \frac{\Delta z}{1 + z} \sim 0.03$$
rearrange the function as $ \Delta z \sim 0.03(1 + z) $. In this case, $ z = z_{UM} = .37 $. Plugging in yields $\Delta z \sim 0.04$. From this, I can use `cosmo.comovingDistance()` to find the $\ell$. Once I have that, I will use 

`np.random.normal(0.0, cosmo.comovingDistance(.37, .37+.04))` 

for the kick, pulling from the distribution defined by the above relations.

```python
# first, get a new table with mocked-up photoz error
um_photoz_err_data = uf.get_mock_photoz_err(um_data, hsc_data)
#um_photoz_err_data.write('um_photoz_err_data.fits')
```

Now find the satellite fraction using this new data set.

```python
uf.open_table('um_photoz_err_sf')
```

```python
um_photoz_err_sf
```

I also want to get a correction factor array for the points, as there is a discrepancy in the calculated values and actual values for satellits fraction in the UM data. This factor will be applied to the HSC data, since we apply the information from the UM tests to the HSC data. I define it as 
$$C = \frac{f_{sat, true}}{f_{sat, alg}}$$
for a given mass bin, such that we get
$$f_{sat, corrected} = Cf_{sat}$$

```python
# find correction factor C
C = um_true['f_sat'] / um_photoz_err_sf['fsat']
# some of the values will have nan because of zero, so
C[4::]=1 # sets no correction where true is zero
```

## 4) Main Plots
There are two main plots to make:
* Results from UM (with/without photoz error)
* HSC results

### 4.1 UM plot

```python
fig = plt.figure(figsize=(8,5))


ax1 = fig.add_subplot(111)
ax1.grid(linewidth=1, linestyle='--', alpha=0.5)

# UM with no photoz error
ax1.errorbar(um_sf['mass'],
             um_sf['fsat'],
             um_sf['err'], 
             color = '#2c7bb6',
             capthick=2, 
             capsize=5, 
             label = 'No photo-z error')

# um with photoz error
ax1.errorbar(um_photoz_err_sf['mass'], 
             um_photoz_err_sf['fsat'], 
             um_photoz_err_sf['err'], 
             color = '#d7191c',
             capthick=2, 
             capsize=5, 
             label = 'With photo-z error')

# add correction
ax1.errorbar(um_photoz_err_sf['mass'], 
             C*um_photoz_err_sf['fsat'], 
             um_photoz_err_sf['err'], 
             color = '#fdae61',
             capthick=2, 
             capsize=5, 
             label = 'With photo-z error + Corrected')

# um true
ax1.plot(um_true['mass'], 
         um_true['f_sat'], 
         ls = '--', 
         color = 'k',
         alpha = 0.7, 
         label = 'UM True')

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(15) 
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(15)



ax1.set_xlabel(r'$ \log \mathrm{M_{\star max} / M_{\odot}}$', fontsize=22)
ax1.set_ylabel(r'$f_{\mathrm{sat}}$ (%)', fontsize=22)
ax1.set_title('$\mathtt{UniverseMachine}$ Satellite Fraction', fontsize = 22)
ax1.legend(fontsize = 12)
```

```python
fig = plt.figure(figsize=(8,5))


ax1 = fig.add_subplot(111)
ax1.grid(linewidth=1, linestyle='--', alpha=0.5)

# hsc
ax1.errorbar(hsc_sf['mass'],
             hsc_sf['fsat'],
             hsc_sf['err'], 
             color = '#d7191c',
             fmt = 's',
             mec = 'k',
             capthick=2, 
             capsize=5, 
             label = 'HSC')

# hsc with correction
ax1.errorbar(hsc_sf['mass'],
             C*hsc_sf['fsat'],
             hsc_sf['err'], 
             color = '#fdae61',
             fmt = 's',
             mec = 'k',
             capthick=2, 
             capsize=5, 
             label = 'HSC + correction')


# um true
ax1.plot(um_true['mass'], 
         um_true['f_sat'], 
         ls = '--', 
         color = 'k',
         alpha = 0.7, 
         label = 'UM True')

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(15) 
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(15)



ax1.set_xlabel(r'$ \log \mathrm{M_{\star max} / M_{\odot}}$', fontsize=22)
ax1.set_ylabel(r'$f_{\mathrm{sat}}$ (%)', fontsize=22)
ax1.set_title('HSC Satellite Fraction', fontsize = 22)
ax1.legend(fontsize = 12)
```
